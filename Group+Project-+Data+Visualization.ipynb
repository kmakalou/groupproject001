# Module 4 - Data Visualization

```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
df = pd.read_csv("nyc-rolling-sales.csv", encoding = "ISO-8859-1") #1st column needs encoding to read pound/hashtag(#) symbol
df.head()
```

```python
#rename the pound/hashtag(#) column so it doesn't cause further issues
df.rename(columns={'#': 'Number'}, inplace =True)
```

```python
#validate
df.head()
```

```python
#make a scatterplot
sns.lmplot(x='RESIDENTIAL UNITS', y='YEAR BUILT', data=df)
```

```python
#boxplot showing stats for each characteristic
plt.subplots(figsize=(13,7))
sns.boxplot(data=df)
```

```python
#create a dataframe containing the stats for each Pokemon
#drop Total, Stage, and Legendary
stats_df = df.drop(['Unnamed: 0', 'BOROUGH', 'NEIGHBORHOOD'], axis=1)
stats_df.head()
```

```python
#boxplot for stats
plt.subplots(figsize=(14,8))
sns.boxplot(data=stats_df)
```

```python
df['ZIP CODE'].dtypes
bins = [10000, 11000, 12000]
stats_df['ZIP CODE'] = pd.cut(stats_df['ZIP CODE'], bins)
df.head()
```

```python
#change size of plot
plt.subplots(figsize=(14,8))

#violin plot shows the distribution of attack power for each type
sns.violinplot(x='ZIP CODE', y='YEAR BUILT', data=stats_df)
```

```python
sns.distplot(df['YEAR BUILT'])
```

```python
#create color palette
#color Hex numbers are from Bulbapedia (https://bulbapedia.bulbagarden.net/wiki/Category:Type_color_templates)
pkmn_type_colors = ['#78C850',  # Grass
                    '#F08030',  # Fire
                    '#6890F0',  # Water
                    '#A8B820',  # Bug
                    '#A8A878',  # Normal
                    '#A040A0',  # Poison
                    '#F8D030',  # Electric
                    '#E0C068',  # Ground
                    '#EE99AC',  # Fairy
                    '#C03028',  # Fighting
                    '#F85888',  # Psychic
                    '#B8A038',  # Rock
                    '#705898',  # Ghost
                    '#98D8D8',  # Ice
                    '#7038F8',  # Dragon
                   ]
```

```python
sns.countplot(x='ZIP CODE',
              data=df, 
              palette=pkmn_type_colors)

plt.xticks(rotation=-45)
```

```python
#swarm plot
plt.subplots(figsize=(10,6))
sns.swarmplot(x='LOT', y='BLOCK', data=df, palette=pkmn_type_colors)
```

```python
#scatterplot by each stage type (evolution level)
g = sns.factorplot(x='ZIP CODE', y='YEAR BUILT',
                   data=df)
                   

#rotate x-axis labels on ALL charts
g.set_xticklabels(rotation=-45)
```

### Visualize all Types by Stat

```python
#create a column that contains all stat types and a column for their corresponding value
melt_df = pd.melt(stats_df, 
                  id_vars=['Name', 'Type 1', 'Type 2'], #column to keep
                  var_name="Stat")

melt_df.head()
```

```python
#melted datframe has 6 times the amount of rows as original stats dataframe
#6 rows for each pokemon for each stat type
print(stats_df.shape)
print(melt_df.shape)
```

```python
plt.figure(figsize=(10,6))

#swarm plot of Stats from melted dataframe
sns.swarmplot(x='Stat', y='value', 
              data=melt_df,
              hue='Type 1')
```

This swarm plot is a bit messy, let's clean it up.

```python
plt.figure(figsize=(13,7))

sns.swarmplot(x='Stat', y='value',
              data=melt_df,
              hue='Type 1', 
              dodge=True, #separate points on chart by hue
              palette=pkmn_type_colors)

#adjust the y-axis
plt.ylim(0,260)

#put the legend on the right
plt.legend(bbox_to_anchor=(1,1), loc=2)
```

```python
#show correlation of stats via heatmap
corr = stats_df.corr()

sns.heatmap(corr, vmin=-1, annot=True)
```

```python
#create a histogram of the distribution of attack power
sns.distplot(df['Grades'])
```

```python
#create a bar plot of Type 1 frequency
sns.countplot(x='Type 1',
              data=df, 
              palette=pkmn_type_colors)

plt.xticks(rotation=-45)
```

```python
#scatterplot by each stage type (evolution level)
g = sns.factorplot(x='Type 1', y='Attack',
                   data=df,
                   hue='Stage', 
                   col='Stage',
                   kind='swarm')

#rotate x-axis labels on ALL charts
g.set_xticklabels(rotation=-45)
```
